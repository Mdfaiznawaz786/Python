6 different patterns generated using Python programming.

```python
# Pattern 1: Right-angled Triangle of Asterisks
def right_triangle(rows):
    for i in range(1, rows + 1):
        print("*" * i)

print("Pattern 1:")
right_triangle(5)
```

```python
# Pattern 2: Inverted Right-angled Triangle of Asterisks
def inverted_right_triangle(rows):
    for i in range(rows, 0, -1):
        print("*" * i)

print("Pattern 2:")
inverted_right_triangle(5)
```

```python
# Pattern 3: Pyramid of Asterisks
def pyramid(rows):
    for i in range(1, rows + 1):
        print(" " * (rows - i) + "* " * i)

print("Pattern 3:")
pyramid(5)
```

```python
# Pattern 4: Diamond of Asterisks
def diamond(rows):
    for i in range(1, rows + 1):
        print(" " * (rows - i) + "* " * i)

    for i in range(rows - 1, 0, -1):
        print(" " * (rows - i) + "* " * i)

print("Pattern 4:")
diamond(5)
```

```python
# Pattern 5: Hollow Square of Asterisks
def hollow_square(size):
    for i in range(size):
        if i == 0 or i == size - 1:
            print("* " * size)
        else:
            print("* " + "  " * (size - 2) + "*")

print("Pattern 5:")
hollow_square(5)
```

```python
# Pattern 6: Hourglass Pattern
def hourglass(rows):
    # Upper part
    for i in range(rows, 0, -1):
        print(" " * (rows - i) + "* " * i)
    # Lower part
    for i in range(2, rows + 1):
        print(" " * (rows - i) + "* " * i)

print("\nPattern 6:")
hourglass(5)
```

Using While Loop

```python
rows = int(input("Enter number of rows: "))
i = 1
while i <= rows:
    j = 0
    while j < i:
        print("*", end=" ")
        j += 1
    print()
    i += 1
```

```python
rows = int(input("Enter number of rows: "))
i = rows
while i > 0:
    j = 0
    while j < i:
        print("*", end=" ")
        j += 1
    print()
    i -= 1
```
