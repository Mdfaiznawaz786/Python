# Mastering Python Lists

Lists are one of the most versatile data structures in Python. Unlike strings, which are immutable, lists are **mutable**, meaning you can modify their contents after they are created.

In this notebook, we will explore:
1. How to create lists
2. Indexing and Slicing
3. Useful List Methods
4. Nesting Lists
5. A quick look at List Comprehensions

```python
# Creating a basic list of integers
prime_numbers = [2, 3, 5, 7, 11]
print(f"List of primes: {prime_numbers}")

# Lists can store multiple data types simultaneously
mixed_data = ['Python', 2024, 3.14, True]
print(f"Mixed data list: {mixed_data}")

# You can check the number of items using len()
print(f"Length of mixed_data: {len(mixed_data)}")
```

### Indexing and Slicing
Lists use zero-based indexing. You can access individual elements or 'slices' of the list just like you do with strings.

```python
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']

# Grab the first item
print(f"First fruit: {fruits[0]}")

# Slice from index 1 to 3 (3 is not included)
print(f"Slice [1:3]: {fruits[1:3]}")

# Grab everything from index 2 onwards
print(f"From index 2: {fruits[2:]}")

# Concatenation (joining lists)
more_fruits = fruits + ['fig', 'grape']
print(f"Concatenated: {more_fruits}")
```

### Essential List Methods
Python provides built-in methods to manipulate lists easily. Note that methods like `.append()` and `.sort()` modify the list **in-place**.

```python
sample_list = [10, 20, 30]

# Add an item to the end
sample_list.append(40)

# Remove and return an item (default is the last index)
last_item = sample_list.pop()
print(f"Popped item: {last_item}")
print(f"List after pop: {sample_list}")

# Sorting and Reversing
alpha_list = ['z', 'a', 'm', 'b']
alpha_list.sort()
print(f"Sorted: {alpha_list}")

alpha_list.reverse()
print(f"Reversed: {alpha_list}")
```

### Nesting Lists
You can place lists inside other lists to create complex structures like matrices.

```python
# A 3x3 matrix
row1 = [1, 2, 3]
row2 = [4, 5, 6]
row3 = [7, 8, 9]

matrix = [row1, row2, row3]

# Accessing the second row
print(f"Second row: {matrix[1]}")

# Accessing the number 6 (second row, third element)
print(f"Specific element: {matrix[1][2]}")
```

### List Comprehensions
List comprehensions provide a concise way to create lists based on existing sequences. It's essentially a one-line `for` loop.

```python
# Extract the first element of every row in our matrix
first_elements = [row[0] for row in matrix]
print(f"First column elements: {first_elements}")

# Create a list of squares
squares = [x**2 for x in range(1, 6)]
print(f"Squares: {squares}")
```
