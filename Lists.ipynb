# Mastering Python Lists

Lists are one of the most versatile data structures in Python. Unlike strings, which are immutable, lists are **mutable**, meaning you can modify their contents after they are created.

In this notebook, we will explore:
1. How to create lists
2. Indexing and Slicing
3. Useful List Methods
4. Nesting Lists
5. A quick look at List Comprehensions

```python
# Creating a basic list of integers
prime_numbers = [2, 3, 5, 7, 11]
print(f"List of primes: {prime_numbers}")

# Lists can store multiple data types simultaneously
mixed_data = ['Python', 2024, 3.14, True]
print(f"Mixed data list: {mixed_data}")

# You can check the number of items using len()
print(f"Length of mixed_data: {len(mixed_data)}")
```

### Indexing and Slicing
Lists use zero-based indexing. You can access individual elements or 'slices' of the list just like you do with strings.

```python
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']

# Grab the first item
print(f"First fruit: {fruits[0]}")

# Slice from index 1 to 3 (3 is not included)
print(f"Slice [1:3]: {fruits[1:3]}")

# Grab everything from index 2 onwards
print(f"From index 2: {fruits[2:]}")

# Concatenation (joining lists)
more_fruits = fruits + ['fig', 'grape']
print(f"Concatenated: {more_fruits}")
```

### Essential List Methods
Python provides built-in methods to manipulate lists easily. Note that methods like `.append()` and `.sort()` modify the list **in-place**.

```python
sample_list = [10, 20, 30]

# Add an item to the end
sample_list.append(40)

# Remove and return an item (default is the last index)
last_item = sample_list.pop()
print(f"Popped item: {last_item}")
print(f"List after pop: {sample_list}")

# Sorting and Reversing
alpha_list = ['z', 'a', 'm', 'b']
alpha_list.sort()
print(f"Sorted: {alpha_list}")

alpha_list.reverse()
print(f"Reversed: {alpha_list}")
```

### Nesting Lists
You can place lists inside other lists to create complex structures like matrices.

```python
# A 3x3 matrix
row1 = [1, 2, 3]
row2 = [4, 5, 6]
row3 = [7, 8, 9]

matrix = [row1, row2, row3]

# Accessing the second row
print(f"Second row: {matrix[1]}")

# Accessing the number 6 (second row, third element)
print(f"Specific element: {matrix[1][2]}")
```

### List Comprehensions
List comprehensions provide a concise way to create lists based on existing sequences. It's essentially a one-line `for` loop.

```python
# Extract the first element of every row in our matrix
first_elements = [row[0] for row in matrix]
print(f"First column elements: {first_elements}")

# Create a list of squares
squares = [x**2 for x in range(1, 6)]
print(f"Squares: {squares}")
```

# Understanding Python Dictionaries

While lists are sequences ordered by position, **Dictionaries** are mappings that store data as **key-value pairs**. If you are coming from other programming languages, you might know these as hash tables or associative arrays.

In this section, we will cover:
1. Creating a Dictionary
2. Accessing and modifying data
3. Nesting Dictionaries
4. Essential Dictionary Methods

```python
# Constructing a dictionary using curly braces {}
# Keys are unique identifiers, values are the associated data
user_info = {'name': 'Alice', 'age': 30, 'city': 'New York'}

# Accessing a value by its key
print(f"User Name: {user_info['name']}")
```

### Flexibility of Dictionaries
Dictionaries are incredibly flexible because the values can be almost any Python object, including integers, strings, or even lists.

```python
data_map = {'key1': 500, 'key2': [10, 20, 30], 'key3': ['apple', 'orange']}

# Accessing a list inside a dictionary
print(f"List at key2: {data_map['key2']}")

# Indexing the list within the dictionary and calling a method
item = data_map['key3'][1].upper()
print(f"Upper case item from list: {item}")
```

### Modifying and Adding Entries
Dictionaries are mutable. You can change the value of an existing key or add entirely new keys through simple assignment.

```python
stats = {'score': 100}

# Modifying a value
stats['score'] = stats['score'] + 50

# Using augmented assignment
stats['score'] += 25
print(f"Final Score: {stats['score']}")

# Adding a new key-value pair to an empty dictionary
registry = {}
registry['event'] = 'Python Workshop'
registry['id'] = 101
print(f"New Registry: {registry}")
```

### Nesting Dictionaries
Just like lists, dictionaries can be nested to create deeply structured data models.

```python
# Deeply nested structure
company = {'dept': {'finance': {'lead': 'Bob', 'budget': 5000}}}

# Grabbing a value deep inside
lead_name = company['dept']['finance']['lead']
print(f"Finance Lead: {lead_name}")
```

### Common Dictionary Methods
There are several built-in methods to view the contents of your dictionary.

```python
my_data = {'k1': 10, 'k2': 20, 'k3': 30}

# Return keys
print(f"Keys: {my_data.keys()}")

# Return values
print(f"Values: {my_data.values()}")

# Return key-value pairs as tuples
print(f"Items: {my_data.items()}")
```

# Python Tuples

In Python, tuples are very similar to lists, however, unlike lists they are **immutable** meaning they cannot be changed. You would use tuples to present things that shouldn't be changed, such as days of the week, or dates on a calendar.

In this section, we will get a brief overview of the following:

1. Constructing Tuples
2. Basic Tuple Methods
3. Immutability
4. When to Use Tuples

```python
# Constructing a tuple using parenthesis ()
t = (1, 2, 3)

# Check len just like a list
print(f"Length of tuple: {len(t)}")

# Can also mix object types
t_mixed = ('one', 2)
print(f"Mixed tuple: {t_mixed}")

# Use indexing just like we did in lists
print(f"First item: {t[0]}")

# Slicing just like a list
print(f"Last item: {t[-1]}")
```

### Basic Tuple Methods

Tuples have built-in methods, but not as many as lists. Let's look at two of them:

```python
t_methods = ('a', 'a', 'b')

# Use .count() to see how many times a value appears
print(f"Count of 'a': {t_methods.count('a')}")

# Use .index() to find the first index of a value
print(f"Index of 'b': {t_methods.index('b')}")
```

### Immutability

It can't be stressed enough, tuples are immutable. To drive that point home:

```python
# Attempting to change a value in a tuple will result in an error
try:
    t[0] = 'change'
except TypeError as e:
    print(f"Error encountered: {e}")

# Because of this immutability, tuples can't grow. Once a tuple is made we can not add to it.
try:
    t.append('nope')
except AttributeError as e:
    print(f"Error encountered: {e}")
```

### When to use Tuples

You may be wondering, "Why bother using tuples when they have fewer available methods?" To be honest, tuples are not used as often as lists in programming, but are used when immutability is necessary. If in your program you are passing around an object and need to make sure it does not get changed, then a tuple becomes your solution. It provides a convenient source of data integrity.

# Python Sets

Sets are an unordered collection of **unique** elements. This means that every item in a set must be distinct. They are constructed using the `set()` function and are particularly useful for membership testing and removing duplicates from other collections.

```python
# Initialize an empty set
my_set = set()

# Adding elements with the .add() method
my_set.add(1)
my_set.add(2)

print(f"Current set: {my_set}")
```

### The Uniqueness Rule
A set will only store one instance of any given value. If you attempt to add an element that is already present, the set will not change.

```python
# Try to add 1 again
my_set.add(1)

# Notice the set remains the same
print(f"Set after duplicate add: {my_set}")
```

### Casting to a Set
A common use case for sets is to quickly find the unique items within a list by 'casting' or converting that list into a set.

```python
# A list with many repeat values
repeated_items = [1, 1, 1, 2, 2, 3, 4, 5, 5, 5, 6]

# Convert to set
unique_set = set(repeated_items)
print(f"Unique values: {unique_set}")
```
